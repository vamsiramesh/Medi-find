```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the sample data
# You can also download the CSV from the link in 1000244290.jpg
url = "https://raw.githubusercontent.com/vamsiramesh/Medi-find/main/Titanic_EDA_Task_02.ipynb" # Example path
# For now, let's use the Titanic data again as it has Age and Sex!
df = sns.load_dataset('titanic')

# 2. Create a Histogram for a Continuous Variable (Age)
plt.figure(figsize=(10, 5))
sns.histplot(df['age'].dropna(), bins=30, kde=True, color='blue')
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# 3. Create a Bar Chart for a Categorical Variable (Gender)
plt.figure(figsize=(8, 5))
sns.countplot(x='sex', data=df, palette='magma')
plt.title('Distribution of Genders')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
